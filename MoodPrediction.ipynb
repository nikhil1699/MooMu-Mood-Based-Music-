```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler 
from sklearn.metrics import classification_report
from sklearn import preprocessing
from sklearn.metrics import accuracy_score
```

```python
#Loading data into dataframe
dataframe = pd.read_csv("sensor_data.csv")
```

```python
#Printing first 5 rows
dataframe.head()
```

```python
#Removing columns with a large number of nan and null values.
dataframe = dataframe[dataframe.columns[dataframe.isnull().mean() < 0.4]]
dataframe = dataframe[dataframe.columns[dataframe.isna().mean() < 0.4]]
```

```python
#Removing null columns
dataframe = dataframe.dropna()
```

```python
columns = np.array(['ACTIVE', 'AFRAID', 'ALERT', 'AMUSED', 'ANGRY', 'ASHAMED', 'ATTENTIVE', 'BORED', 'CALM', 'CRAZY', 'DETERMINED', 'DISGUSTED', 'DISTRESSED', 'DREAMY', 'ENERGETIC', 'ENTHUSIASTIC', 'EXCITED', 'FRUSTRATED', 'GUILTY', 'HAPPY', 'HIGH', 'HOSTILE', 'HUNGRY', 'IN_EMOTIONAL_PAIN', 'IN_LOVE', 'IN_PHYSICAL_PAIN', 'INSPIRED', 'INTERESTED', 'IRRITABLE', 'JITTERY', 'LONELY', 'NERVOUS', 'NORMAL', 'NOSTALGIC', 'OPTIMISTIC', 'PHYSICALLY_SICK', 'PROUD', 'ROMANTIC', 'SAD', 'SCARED', 'SERIOUS', 'SEXY', 'SLEEPY', 'STRESSED', 'STRONG', 'TIRED', 'UNTROUBLED', 'UPSET', 'WORRIED'])
columns.size
```

```python
dataframe.shape
```

```python
#Allocating features and target
features = dataframe.iloc[:, 0:211]
target  = dataframe.iloc[: , -49:]
```

```python
#Dropping target variables with no positive samples
target = target.drop(columns = ['AFRAID', 'CRAZY', 'DISGUSTED', 'DISTRESSED', 'GUILTY', 'JITTERY', 'PROUD', 'ROMANTIC', 'SCARED', 'STRONG'])
target.columns.size
```

```python
#Neural Network
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.30) 
scaler = StandardScaler()
scaler.fit(X_train)  
X_train = scaler.transform(X_train) 
X_test = scaler.transform(X_test)  
clf = MLPClassifier(solver='adam', alpha=1e-5, hidden_layer_sizes=(50, 15), random_state=1, 
                    verbose = "true", max_iter = 1000)
clf.fit(X_train, y_train)
```

```python
#Creating buckets for emotion.
def get_emotion(s):
    if s == 'ATTENTIVE' or s == 'BORED' or s == 'CALM' or s == 'NORMAL' or s == 'HIGH' or s == 'SERIOUS' or s == 'SLEEPY' or s == 'TIRED' or s == 'UNTROUBLED':
        return 'NEUTRAL'
    elif s == 'ACTIVE' or s == 'DETERMINED' or s == 'ENERGETIC' or s == 'AMUSED' or s == 'ENTHUSIASTIC' or s == 'EXCITED' or s == 'HAPPY' or s == 'INSPIRED' or s == 'INTERESTED':
        return 'HAPPY'
    elif s == 'ASHAMED' or s == 'IN_PHYSICAL_PAIN' or s == 'PHYSICALLY_SICK':
        return 'FEAR'
    elif s == 'ALERT':
        return 'SURPRISED'
    elif s == 'DREAMY' or s == 'IN_LOVE' or s == 'SEXY':
        return 'ROMANTIC'
    elif s == 'ANGRY' or s == 'HOSTILE' or s == 'HUNGRY' or s == 'IRRITABLE':
        return 'ANGRY'
    elif s == 'WORRIED' or s == 'STRESSED' or s == 'NOSTALGIA' or s == 'NERVOUS' or s == 'SAD' or s == 'FRUSTRATED' or s == 'LONELY' or s == 'IN_EMOTIONAL_PAIN' or s == 'UPSET':
        return 'SAD'
```

```python
from sklearn.multioutput import MultiOutputClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn import tree
import xgboost as xgb
```

```python
#Logistic Regression
clf = MultiOutputClassifier(LogisticRegression(max_iter = 1000)).fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
print("Accuracy", accuracy_score(y_test, y_pred))
```

```python
#Gaussian Naive Bayes
clf = MultiOutputClassifier(GaussianNB()).fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
print("Accuracy", accuracy_score(y_test, y_pred))
```

```python
#Decision Tree
clf = MultiOutputClassifier(tree.DecisionTreeClassifier()).fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
print("Accuracy", accuracy_score(y_test, y_pred))
```

```python
#XGBoost
clf = MultiOutputClassifier(xgb.XGBClassifier(tree_method='hist')).fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
print("Accuracy", accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import multilabel_confusion_matrix
#Printing confusion matrix for XGBoost
multilabel_confusion_matrix(y_test, y_pred)
```
