```python
def get_word_mapping(le):
    res = {}
    for cl in le.classes_:
        res.update({le.transform([cl])[0]:cl})

    return res

def get_integer_mapping(le):
    res = {}
    for cl in le.classes_:
        res.update({cl:le.transform([cl])[0]})

    return res
```
